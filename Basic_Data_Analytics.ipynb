Getting cleaner descriptions and a spread variable

```python
user_ads = pd.read_csv("ads_dataset.csv")

# Creates a new data frame called desc_data tha describes the ads dataframe
desc_data = user_ads.describe()

# Exchange rows and columns in pandas
desc_data = desc_data.transpose()

# Adds a column called spread that is the result of subtracting the min from the max.
desc_data.loc[:, "spread"] = desc_data["max"] - desc_data["min"]
desc_data
```

Finding the features that have a spread of 1

```python
binary = desc_data[desc_data["spread"]==1]
binary
```
